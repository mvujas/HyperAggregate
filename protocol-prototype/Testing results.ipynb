```python
x = [3, 6, 9, 12, 15, 21, 30]

time_full = [
    data[i]['full']['time']\
    for i in x
]
time_small = [
    data[i]['small']['time']\
    for i in x
]

time_coef = [
    time_full[i] / time_small[i]\
    for i in range(len(time_full))
]

    



plt.figure(figsize=(15, 10), dpi=80)
line_full, = plt.plot(x, time_full, label='All peers in the same aggregation [case A]')
line_small, = plt.plot(x, time_small, label='3 peers per single aggregation [case B]')
line_rel, = plt.plot(x, time_coef, label='Relative time [A / B]')
plt.ylabel('Average time per aggregation (seconds)')
plt.xlabel('Number of peers')
plt.title('Change of time per aggregation depending on the number of participants')
plt.legend(handles=[line_full, line_small, line_rel])
plt.show()
```
